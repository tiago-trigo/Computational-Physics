```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/sunspots.txt", sep="	", names=["Month","Sunspots"])
```

```python
fig, ax  = plt.subplots()

ax.plot(df["Month"][:1001], df["Sunspots"][:1001], linewidth = 1)
```

```python
rolling = []

for i in range(5,1005):
    value = 1/11 * sum([df["Sunspots"][index] for index in range(i-5, i+5)])
    rolling.append(value)
```

```python
fig, ax  = plt.subplots()

ax.plot(df["Month"][:1001], df["Sunspots"][:1001], linewidth = 1)
ax.plot(df["Month"][:1000], rolling, linewidth = 2)
```

```python
t = np.linspace(0, 2*np.pi, 1000)
x = 2 * np.cos(t) + np.cos(2*t)
y = 2 * np.sin(t) - np.sin(2*t)
```

```python
fig, ax  = plt.subplots()

ax.plot(x,y)
```

```python
t = np.linspace(0, 10*np.pi, 1000)
r = t**2

fig, ax  = plt.subplots(subplot_kw={'projection': 'polar'})

ax.plot(t, r)
```

```python
def polar_conv(t, r):
    x = r * np.cos(t)
    y = r * np.sin(t)
    
    return x, y

fig, ax = plt.subplots(figsize = (6,6))

ax.plot(x,y)
```

```python
t = np.linspace(0, 24*np.pi, 1000)
r = np.exp(np.cos(t)) - 2 * np.cos(4*t) + (np.sin(t/12))**5

x,y = polar_conv(t,r)

fig, ax = plt.subplots(figsize = (6,6))

ax.plot(x,y)
```
